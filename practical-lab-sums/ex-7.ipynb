```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Generate two groups of points
np.random.seed(42)
group1 = np.random.normal(loc=2, scale=1, size=(50, 2))  
group2 = np.random.normal(loc=7, scale=1, size=(50, 2))  
data = np.vstack([group1, group2]) 


gmm = GaussianMixture(n_components=2) 
gmm.fit(data)
labels = gmm.predict(data)

# Plot the results
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=30)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='x', s=100)  # Cluster centers
plt.title("Clusters Found by EM")
plt.show()
```
